```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
from matplotlib.patches import Rectangle
from matplotlib import cm
from matplotlib import rc as matplotlibrc
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import  MultipleLocator
import pandas as pd
matplotlibrc('text',usetex=True)
matplotlibrc('font',family='serif')
```

```python
Upperlimit_comp  = pd.read_csv('UpperLimit_comp.csv')
Lowerlimit_comp  = pd.read_csv('LowerLimit_comp.csv')

Upperlimit_expr  = pd.read_csv('UpperLimit_exp.csv')
Lowerlimit_expr  = pd.read_csv('LowerLimit_exp.csv')
MidRange_expr    = pd.read_csv('MidRange_exp.csv')
```

```python
plt.figure(figsize=(11,6))


plt.plot(np.array(Upperlimit_comp['M']),np.array(Upperlimit_comp['l']),'s--',mfc='w',markersize=10,mec='k',color='k',label='Comp. upper limit')
plt.plot(np.array(Lowerlimit_comp['M']),np.array(Lowerlimit_comp['l']),'o--',mfc='w',markersize=10,mec='k',color='k',label='Comp. lower limit')

plt.plot(np.array(Upperlimit_expr['M']),np.array(Upperlimit_expr['l']),'s',markersize=10,mec='k',mfc='k',label='Exp. Collective shock')
plt.plot(np.array(Lowerlimit_expr['M']),np.array(Lowerlimit_expr['l']),'o',markersize=10,mec='k',mfc='k',label='Exp. Regular int.')
plt.plot(np.array(MidRange_expr['M']),np.array(MidRange_expr['l']),'^',markersize=10,mec='k',mfc='k',label='Exp. Mach conf.')


plt.xlabel('$Ma_\infty$',fontsize=22)
plt.ylabel('$\lambda$',fontsize=22)
plt.tick_params(axis='both',which='both',labelsize=18, direction = 'in')

####### INTERPOLATION ############################################
N  = 1.5+(np.array(Lowerlimit_comp['l'])[-1]-np.array(Lowerlimit_comp['l'])[-2])/(np.array(Lowerlimit_comp['M'])[-1]-np.array(Lowerlimit_comp['M'])[-2])*2.0
NN = np.array(Upperlimit_comp['l'])[-1]+(np.array(Lowerlimit_comp['l'])[-1]-np.array(Lowerlimit_comp['l'])[-2])/(np.array(Lowerlimit_comp['M'])[-1]-np.array(Lowerlimit_comp['M'])[-2])*2.0

plt.plot((2,4),(1.5,N),'b',label='Interpolation',marker='x',markersize=10)
plt.plot((2,4),(np.array(Upperlimit_comp['l'])[-1],NN),'b',marker='x',markersize=10)

plt.plot((2,3,4),(1.5,1.5,1.5),'*',markersize=12,mfc='w',mec='r',label='DSMC')

plt.legend(frameon=False,fontsize=18)

#plt.ylim(0,13)
plt.savefig('Fomin.pdf')
```
